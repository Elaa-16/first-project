```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. Charger les donn√©es XOR depuis le fichier CSV
df = pd.read_csv('xor.csv')  # Le fichier doit contenir les colonnes 'x1', 'x2' et 'label'

# 2. S√©parer les entr√©es (X) et les labels (y)
X = df[['x1', 'x2']]
y = df['label']

# 3. S√©parer en jeu d'entra√Ænement et de test (ici 50%/50% √† cause de la petite taille)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42, stratify=y)

# 4. D√©finir le mod√®le MLP
mlp = MLPClassifier(
    hidden_layer_sizes=(4,),  # 1 couche cach√©e avec 4 neurones
    activation='relu',
    max_iter=3000,
    random_state=1,
    # Pas d'early_stopping ici
)

# 5. Entra√Æner le mod√®le sur les donn√©es d'entra√Ænement
mlp.fit(X_train, y_train)

# 6. Pr√©dire sur le jeu de test
y_pred = mlp.predict(X_test)

# 7. √âvaluer les performances
print("\n‚úÖ Accuracy sur le test :", accuracy_score(y_test, y_pred))
print("\nüßæ Rapport de classification sur le test :\n", classification_report(y_test, y_pred))

# 8. Tracer la courbe de perte pendant l'entra√Ænement
plt.plot(mlp.loss_curve_)
plt.title("Courbe de perte du MLP")
plt.xlabel("It√©rations")
plt.ylabel("Perte (loss)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

# 1. Charger les donn√©es XOR depuis le fichier CSV
df = pd.read_csv('xor.csv')  # Le fichier doit contenir x1, x2 et label

# 2. S√©parer les entr√©es (X) et les labels (y)
X = df[['x1', 'x2']]
y = df['label']

# 3. D√©finir le mod√®le MLP
mlp = MLPClassifier(
    hidden_layer_sizes=(4,),
    activation='relu',
    max_iter=3000,
    random_state=1,
    verbose=True
    # ‚ö†Ô∏è PAS de early_stopping ici car trop peu de donn√©es
)

# 4. Entra√Æner le mod√®le sur toutes les donn√©es
mlp.fit(X, y)

# 5. √âvaluer le mod√®le sur les m√™mes donn√©es (puisqu‚Äôon n‚Äôa pas de test set ici)
y_pred = mlp.predict(X)
print("\n‚úÖ Accuracy:", accuracy_score(y, y_pred))
print("\nüßæ Classification Report:\n", classification_report(y, y_pred))

# 6. Tracer la courbe de perte
plt.plot(mlp.loss_curve_)
plt.title("Courbe de perte du MLP")
plt.xlabel("It√©rations")
plt.ylabel("Perte (loss)")
plt.grid(True)
plt.show()

```

```python
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

df = pd.read_csv('xor.csv')
X = df[['x1', 'x2']]
y = df['label']

mlp = MLPClassifier(
    hidden_layer_sizes=(4,),
    activation='tanh',
    max_iter=3000,
    random_state=1,
)

mlp.fit(X, y)
y_pred = mlp.predict(X)

print("Accuracy sur tout le dataset :", accuracy_score(y, y_pred))
print("Rapport de classification :\n", classification_report(y, y_pred))

plt.plot(mlp.loss_curve_)
plt.title("Courbe de perte du MLP")
plt.xlabel("It√©rations")
plt.ylabel("Perte (loss)")
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Load data
df = pd.read_csv("xor_3.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Scale input data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create and train a deeper MLP
mlp = MLPClassifier(hidden_layer_sizes=(10, 5), activation='tanh', max_iter=5000, random_state=42)
mlp.fit(X_scaled, y)

# Predict
preds = mlp.predict(X_scaled)
acc = accuracy_score(y, preds)

print("Predictions:", preds)
print(f"Accuracy: {acc:.2f}")
```
